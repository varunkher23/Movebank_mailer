```python
# Ensure the key columns are named consistently
df_branch = df_branch.rename(columns={"ec5_branch_owner_uuid": "ec5_uuid"})

# Perform the left join
df = df_lastweek.merge(df_branch, on="ec5_uuid", how="left")

df = df[[
    "3_Date", "4_Start_Time", "5_Type_of_survey", "2_Name_of_the_person",
    "8_Species", "13_No_of_adult_males", "14_No_of_adult_femal",
    "15_No_of_immature_fe", "16_No_of_immature_ma", "17_No_of_unknown_age",
    "18_Notes", "20_Location", "21_Distance_from_obs", "22_Angle",
    "34_End_time", '35_Areas_covered','36_Total_Kms_covered'
]]


# Ensure time columns are in datetime format
df["4_Start_Time"] = pd.to_datetime(df["4_Start_Time"], errors="coerce")
df["34_End_time"] = pd.to_datetime(df["34_End_time"], errors="coerce")

# Calculate time spent in minutes
df["time_spent"] = ((df["34_End_time"] - df["4_Start_Time"]).dt.total_seconds() / 60).round(2)

# Ensure numeric columns are correct
df["13_No_of_adult_males"] = pd.to_numeric(df["13_No_of_adult_males"], errors="coerce")
df["14_No_of_adult_femal"] = pd.to_numeric(df["14_No_of_adult_femal"], errors="coerce")
df["15_No_of_immature_fe"] = pd.to_numeric(df["15_No_of_immature_fe"], errors="coerce")
df["16_No_of_immature_ma"] = pd.to_numeric(df["16_No_of_immature_ma"], errors="coerce")
df["17_No_of_unknown_age"] = pd.to_numeric(df["17_No_of_unknown_age"], errors="coerce")
df["36_Total_Kms_covered"] = pd.to_numeric(df["36_Total_Kms_covered"], errors="coerce")

# Add a column for period of day
df["survey_period"] = df["4_Start_Time"].dt.hour.apply(lambda x: "Morning" if x < 12 else "Evening")

# Main summary: include survey_period in groupby
df_summary = df.groupby(["3_Date", "survey_period"]).agg({
    "2_Name_of_the_person": lambda x: ", ".join(sorted(set(
        name for sublist in x for name in (sublist if isinstance(sublist, list) else [sublist])
    ))),
    "35_Areas_covered": lambda x: ", ".join(sorted(set(
        name for sublist in x for name in (sublist if isinstance(sublist, list) else [sublist])
    ))),
    "14_No_of_adult_femal": "sum",
    "13_No_of_adult_males": "sum",
    "15_No_of_immature_fe": "sum",
    "16_No_of_immature_ma": "sum",
    "17_No_of_unknown_age": "sum",
    "36_Total_Kms_covered": "sum"
}).reset_index()

# Add bird totals
df_summary["Immature_birds"] = (
    df_summary["15_No_of_immature_fe"] +
    df_summary["16_No_of_immature_ma"] +
    df_summary["17_No_of_unknown_age"]
)
df_summary["Total_birds"] = (
    df_summary["14_No_of_adult_femal"] +
    df_summary["13_No_of_adult_males"] +
    df_summary["Immature_birds"]
)

count_cols = ["Total_birds","Immature_birds",
    "14_No_of_adult_femal", "13_No_of_adult_males",
    "15_No_of_immature_fe", "16_No_of_immature_ma", "17_No_of_unknown_age"
]

df_summary[count_cols] = df_summary[count_cols].fillna(0).astype(int)

# Summarize time spent by date, period, and survey type
time_by_type = df.groupby(["3_Date", "survey_period", "5_Type_of_survey"])["time_spent"].sum().unstack(fill_value=0).reset_index()

# Ensure columns name is clean
time_by_type.columns.name = None

# Add missing columns if not present
if "Vantage" not in time_by_type.columns:
    time_by_type["Vantage"] = 0

if "Vehicle" not in time_by_type.columns:
    time_by_type["Vehicle"] = 0

# Rename for clarity
time_by_type = time_by_type.rename(columns={
    "3_Date": "3_Date",
    "Vantage": "Time_Spent_Vantage",
    "Vehicle": "Time_Spent_Vehicle"
})

# Merge
df_summary = df_summary.merge(time_by_type, on=["3_Date", "survey_period"], how="left")

# Rename and reorder columns
df_summary = df_summary.rename(columns={
    "3_Date": "Date",
    "35_Areas_covered": "Areas covered",
    "2_Name_of_the_person": "Observers",
    "14_No_of_adult_femal": "Females",
    "13_No_of_adult_males": "Males",
    "36_Total_Kms_covered": "Total_Km_Surveyed",
    "survey_period": "Period"
})

df_summary = df_summary[[
    "Date", "Period", "Areas covered","Observers", "Total_birds", "Females", "Males",
    "Immature_birds", "Time_Spent_Vantage", "Time_Spent_Vehicle", "Total_Km_Surveyed"
]]

df_summary_html=df_summary.to_html()
```

```python
import numpy as np
import pandas as pd
from pyproj import Proj, Transformer

# Extract fields from 20_Location dict
df["latitude"] = df["20_Location"].apply(lambda loc: loc.get("latitude") if isinstance(loc, dict) else None)
df["longitude"] = df["20_Location"].apply(lambda loc: loc.get("longitude") if isinstance(loc, dict) else None)
df["UTM_Easting"] = df["20_Location"].apply(lambda loc: loc.get("UTM_Easting") if isinstance(loc, dict) else None)
df["UTM_Northing"] = df["20_Location"].apply(lambda loc: loc.get("UTM_Northing") if isinstance(loc, dict) else None)
df["UTM_Zone"] = df["20_Location"].apply(lambda loc: loc.get("UTM_Zone") if isinstance(loc, dict) else None)

# Create df_locations with relevant fields
df_locations = df[[
    "3_Date", "4_Start_Time", "5_Type_of_survey", "2_Name_of_the_person",
    "8_Species", "13_No_of_adult_males", "14_No_of_adult_femal",
    "15_No_of_immature_fe", "16_No_of_immature_ma", "17_No_of_unknown_age",
    "latitude", "longitude", "UTM_Easting", "UTM_Northing", "UTM_Zone"
]].copy()

# Ensure necessary columns are numeric
df["UTM_Easting"] = pd.to_numeric(df["UTM_Easting"], errors="coerce")
df["UTM_Northing"] = pd.to_numeric(df["UTM_Northing"], errors="coerce")
df["21_Distance_from_obs"] = pd.to_numeric(df["21_Distance_from_obs"], errors="coerce")
df["22_Angle"] = pd.to_numeric(df["22_Angle"], errors="coerce")

# Convert angle to radians
df["angle_rad"] = np.deg2rad(df["22_Angle"])

# Calculate UTM offsets
df["delta_E"] = df["21_Distance_from_obs"]/1000 * np.sin(df["angle_rad"])
df["delta_N"] = df["21_Distance_from_obs"]/1000 * np.cos(df["angle_rad"])

# Projected UTM coordinates
df["proj_UTM_Easting"] = df["UTM_Easting"] + df["delta_E"]
df["proj_UTM_Northing"] = df["UTM_Northing"] + df["delta_N"]

# Fixing hemisphere logic
# Function to convert UTM Zone string like "42R" into EPSG code
def utm_zone_to_epsg(zone):
    number = int(zone[:-1])
    hemisphere = zone[-1].upper()
    return 32600 + number if hemisphere >= 'N' else 32700 + number
    
# Safe conversion function
def safe_convert_utm_to_latlon(row):
    try:
        epsg_code = utm_zone_to_epsg(row["UTM_Zone"])
        transformer = Transformer.from_crs(f"EPSG:{epsg_code}", "EPSG:4326", always_xy=True)
        lon, lat = transformer.transform(row["proj_UTM_Easting"], row["proj_UTM_Northing"])
        return pd.Series([lat, lon])
    except Exception as e:
        # You can log the error here if needed
        return pd.Series([None, None])

# Apply to DataFrame
df[["proj_latitude", "proj_longitude"]] = df.apply(safe_convert_utm_to_latlon, axis=1)

# Final projected location DataFrame
df_projected_locations = df[[
    "3_Date", "4_Start_Time", "5_Type_of_survey", "2_Name_of_the_person",
    "8_Species", "13_No_of_adult_males", "14_No_of_adult_femal",
    "15_No_of_immature_fe", "16_No_of_immature_ma", "17_No_of_unknown_age",
    "18_Notes", "20_Location", "21_Distance_from_obs", "22_Angle",
    "34_End_time","proj_latitude", "proj_longitude"
]]

# Ensure the '4_Start_Time' column is in datetime format
df_projected_locations["4_Start_Time"] = pd.to_datetime(df_projected_locations["4_Start_Time"], errors="coerce")

# Extract only the time part
df_projected_locations["4_Start_Time"] = df_projected_locations["4_Start_Time"].dt.time

# Optionally display or export
print(df_projected_locations.head())
```

```python
import simplekml

# Create a new KML object
kml = simplekml.Kml()

#df_projected_locations["proj_latitude"] = df_projected_locations["proj_latitude"].map(lambda x: f"{x:.5f} N")
#df_projected_locations["proj_longitude"] = df_projected_locations["proj_longitude"].map(lambda x: f"{x:.5f} E")
df_projected_locations["proj_latitude"].replace('', np.nan, inplace=True)
df_projected_locations["proj_longitude"].replace('', np.nan, inplace=True)
df_projected_locations = df_projected_locations.dropna(subset=['proj_latitude', 'proj_longitude'])
# Loop through each row in the projected location DataFrame
for _, row in df_projected_locations.iterrows():
    species = row["8_Species"]
    observer = row["2_Name_of_the_person"]
    date = row["3_Date"]
    lat = row["proj_latitude"]
    lon = row["proj_longitude"]
    
    # Add a placemark
    kml.newpoint(
        name=str(species),
        coords=[(lon, lat)],
        description=f"Observer: {observer}\nDate: {date}"
    )

# Save the KML file
kml.save("projected_sightings.kml")
```
